```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
train = pd.read_csv("train.csv")
train
```

```python
test = pd.read_csv("test.csv")
test
```

```python
train_price = train['SalePrice']

train_price # our Y array

input_data=train[['OverallQual', 'GrLivArea',	'GarageCars',	'GarageArea',	'TotalBsmtSF',	'1stFlrSF',	'FullBath',	'TotRmsAbvGrd',	'YearBuilt',	'YearRemodAdd']]
```

```python
test_price = test['SalePrice']
test_price
```

```python
y = np.array(train_price)
x = np.array(input_data)
x
```

```python
linreg = LinearRegression()
```

```python
y = y.reshape(-1,1)
y.shape
y
```

```python
y = y.reshape(-1,1)
y.shape
```

```python
linreg.fit(x,y)
```

```python
print(linreg.coef_)
```

```python
test_x=np.array(test[['OverallQual', 'GrLivArea',	'GarageCars',	'GarageArea',	'TotalBsmtSF',	'1stFlrSF',	'FullBath',	'TotRmsAbvGrd',	'YearBuilt',	'YearRemodAdd']])
```

```python
predictions=linreg.predict(test_x)
predictions
```

```python
y_test=test['SalePrice']
r2 = r2_score(y_test, predictions)

print(f"R-squared on Validation Set: {r2:.2f}")
y_test
plt.scatter(y_test, predictions)

```
